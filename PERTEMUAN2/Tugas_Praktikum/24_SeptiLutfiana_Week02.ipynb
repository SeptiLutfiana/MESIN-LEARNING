# 4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
import pandas as pd

df2 = pd.read_csv('E:\KULIAH\SEM 5\MESIN LEARNING\PERTEMUAN2\data\wbc.csv')
```

```python
# Split data
from sklearn.model_selection import train_test_split

# Split data training dan dan lainnya
# data lainnya, akan kita split lagi menjadi validasi dan testing.
# Rasio yang akan kita gunakan adalah 8:1:1
df2_train, df2_test = train_test_split(df2, test_size=0.2, stratify=df2['diagnosis'])

# Cek masing-masing ukuran data

print(f'Jumlah label data asli:\n{df2.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df2_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df2_test.diagnosis.value_counts()}')
```
